```python
df.dropna(inplace=True)  # remove empty rows
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")  # clean column names
```

```python
from textblob import TextBlob

def get_sentiment(text):
    if not isinstance(text, str) or text.strip() == "":
        return 0  # neutral for blanks
    return TextBlob(text).sentiment.polarity

# Apply sentiment to each feedback column (customize for your dataset)
feedback_cols = ['teaching', 'coursecontent', 'examination', 'labwork', 'library_facilities', 'extracurricular']

for col in feedback_cols:
    df[f'{col}_sentiment'] = df[col].apply(get_sentiment)
```

```python
def label_sentiment(score):
    if score > 0.1:
        return "Positive"
    elif score < -0.1:
        return "Negative"
    else:
        return "Neutral"

for col in feedback_cols:
    df[f'{col}_label'] = df[f'{col}_sentiment'].apply(label_sentiment)
```

```python
import pandas as pd
from textblob import TextBlob

# Load your dataset
df = pd.read_csv("Copy of finalDataset0.2.csv")  # replace with your actual file path

# Define sentiment function
def get_sentiment(text):
    polarity = TextBlob(str(text)).sentiment.polarity
    if polarity > 0.1:
        return 1
    elif polarity < -0.1:
        return -1
    else:
        return 0

# List of columns to analyze
feedback_columns = ['teaching', 'coursecontent', 'Examination', 'labwork', ' library_facilities', 'extracurricular']

# Add sentiment columns beside each feedback column
for col in feedback_columns:
    sentiment_col = col + '_sentiment'
    df[sentiment_col] = df[col].apply(get_sentiment)

# Optional: Add a "total_sentiment" column per row
df['total_sentiment'] = df[[col + '_sentiment' for col in feedback_columns]].sum(axis=1)

# Optional: Add an "average_sentiment" column per row
df['average_sentiment'] = df[[col + '_sentiment' for col in feedback_columns]].mean(axis=1)

# Show the final result
print(df.head())

# Save to Excel or CSV if needed
df.to_csv("sentiment_output.csv", index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

summary = {}

for col in feedback_cols:
    summary[col] = df[f'{col}_label'].value_counts()

sentiment_df = pd.DataFrame(summary).T
sentiment_df.plot(kind='bar', stacked=True, figsize=(12, 6), colormap="Set2")
plt.title("Sentiment Comparison Across Feedback Topics")
plt.ylabel("Number of Responses")
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns

heat_data = df[[col + '_sentiment' for col in feedback_columns]]
sns.heatmap(heat_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Feedback Sentiments')
plt.show()
```

```python
sns.boxplot(data=df[[col + '_sentiment' for col in feedback_columns]])
plt.title('Sentiment Distribution in Each Feedback Category')
plt.xticks(rotation=45)
plt.ylabel('Sentiment Score (-1 to 1)')
plt.show()
```

```python
from wordcloud import WordCloud

for col in feedback_cols:
    text = " ".join(df[col].dropna().astype(str))
    wordcloud = WordCloud(background_color='white', max_words=100).generate(text)
    plt.figure(figsize=(8,4))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f"Word Cloud: {col.capitalize()}")
    plt.show()
```

**Campus Feedback Analysis: Internship Project Summary**
As part of my internship project, I conducted a comprehensive analysis of student feedback collected through Google Forms for various campus experiences. This dataset included both structured ratings and open-text responses, enabling a holistic understanding of student sentiment across academics, labs, exams, and extracurricular activities.

Using Python (pandas, matplotlib, seaborn) for data processing and TextBlob for sentiment analysis, I cleaned and structured the data to uncover actionable insights. Sentiment analysis revealed that over 70% of students expressed positive feedback regarding teaching, especially praising punctuality, clarity, and instructor engagement. However, course content received largely neutral sentiment, indicating the need for more interactive or updated material to enhance academic delivery.

When analyzing feedback on exams, student responses were highly varied. Many appreciated fairness and clarity, while others expressed concern over inconsistent difficulty. This suggested a need for balancing complexity across assessments. Lab sessions were noted as essential but often criticized as not up to the markprompting a recommendation for improved scheduling and practical depth.

On a positive note, extracurricular activities stood out with very high sentiment scores, with students appreciating the diversity and quality of non-academic offerings. These programs play a crucial role in boosting student morale and campus life satisfaction.

**Key Recommendations:**
Revise lab structure and practical exposure to improve student satisfaction.

Update course materials to make academic content more engaging and relevant.

Review exam patterns to ensure fair and balanced evaluation.

Continue supporting a wide range of extracurricular initiatives.
